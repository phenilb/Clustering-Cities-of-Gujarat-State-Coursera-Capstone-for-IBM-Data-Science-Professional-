```python
import pandas as pd
import folium
import numpy as np
```

```python
df = pd.read_csv("gujsemicompiled.csv")
df2 = pd.read_csv("gujcityvenuecounts.csv")

df.head()
```

```python
df2.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
df2.drop("Unnamed: 0", axis=1, inplace=True)
df2.head()
```

```python
df_merged = pd.merge(df,df2,on="City", how="inner")
df_merged
```

```python
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

```python
df_merged.columns
```

```python
df_merged = df_merged[['City', 'Latitude', 'Longitude', \
'Population', 'Population_Density', 'Total Talukas', 'UrbanPopPercentage', 'GrowthRate', 'SexRatio',\
'LiteracyRate', 'Total Schools', 'Teachers', 'SchoolsWithPlaygrounds','SchoolsWithComputers', 'Venue Counts']]

df_merged.head()
```

```python
df_merged.describe()
```

```python
df_merged['Population_Density'].dtype
```

```python
df_merged['Population_Density'] = df_merged['Population_Density'].replace({',': ''}, regex=True)
df_merged['Population_Density'] = df_merged['Population_Density'].astype(float)
df_merged
```

```python
X = df_merged.values[:,3:].astype(float)
X_scaled = StandardScaler().fit_transform(X)
X_scaled
```

```python
import matplotlib.pyplot as plt # plotting library

from sklearn.cluster import KMeans 

num_clusters = 5

k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=20)
k_means.fit(X_scaled)
labels = k_means.labels_

print(labels)
```

```python
df_kmeans = df_merged
df_kmeans.head()
```

```python
df_kmeans['Cluster'] = labels
```

```python
df_kmeans.head()
```

```python
df_kmeans.groupby('Cluster').mean()
```

```python
from sklearn.cluster import AgglomerativeClustering 

agglom = AgglomerativeClustering(n_clusters = 3, linkage = 'complete')
agglom.fit(X_scaled)
alabels = agglom.labels_
alabels
```

```python
df_agglo = df_merged
df_agglo['Cluster_Agglo'] = alabels
df_agglo.head()
```

```python
df_agglo.to_pickle("FinalDataFrame.pickle")
df_agglo.to_csv("FinalDataFrame.csv")
```
