```python
import pandas as pd

df = pd.read_csv("citieswithloc.csv")
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

```python
df2 = pd.read_csv("edudata.csv")
df2.head()
```

```python
df2.drop('Unnamed: 0', axis=1, inplace=True)
df2.head()
```

```python
df['City'] = df['City'].apply(lambda x: x.upper())
df.head()
```

```python
from bs4 import BeautifulSoup
import requests
website_html = requests.get('https://en.wikipedia.org/wiki/List_of_districts_of_Gujarat').text
soup = BeautifulSoup(website_html, 'lxml')

mylist = []
for tr in soup.find_all('tr')[1:]:
    tds = tr.find_all('td')
    try:
        mylist.append([tds[0].text.rstrip(), tds[1].text.rstrip(), tds[2].text.rstrip(),\
                       tds[3].text.rstrip(), tds[4].text.rstrip(), tds[5].text.rstrip(),\
                       tds[6].text.rstrip(), tds[7].text.rstrip(), tds[8].text.rstrip(), tds[9].text.rstrip()])    
    except:
        next
df3 = pd.DataFrame(mylist, columns=['No.', 'District', 'Capital', \
                                   'Population 2001 Census', 'Population 2011 Census', 'Area', \
                                   'Density 2011', 'Year of Formation', 'Taluka', 'Total Talukas'])

df3.drop(['No.', 'Population 2001 Census', 'Area', 'Year of Formation', 'Taluka'], axis=1, inplace=True)
df3.head(7)
```

```python
df3['District'] = df3['District'].apply(lambda x: x.upper())
df3
```

```python
df_merged = pd.merge(df2,df3,on='District',how='inner')
df_merged.head(7)
```

```python
df_merged.drop('District',axis=1,inplace=True)
df_merged
```

```python
df_merged.columns
```

```python
df_merged = df_merged[['Capital', 'UrbanPopPercentage', 'GrowthRate', 'SexRatio', 'LiteracyRate',
       'Total Schools', 'Teachers', 'SchoolsWithPlaygrounds',
       'SchoolsWithComputers', 'Population 2011 Census',
       'Density 2011', 'Total Talukas']]
df_merged
```

```python
df_merged.dropna(axis=0, inplace=True)
df_merged
```

```python
df_merged.rename(columns={'Capital':'City'},inplace=True)
df_merged['City'] = df_merged['City'].apply(lambda x: x.upper())
df_merged.head()
```
