```python
import pandas as pd
import folium
import matplotlib.pyplot as plt

%matplotlib inline

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering 
```

```python
df = pd.read_csv("SemiFinalDF.csv")
df
```

```python
df.columns
```

```python
df = df[['City', 'Latitude', 'Longitude', 'Population','GrowthRate', 'SexRatio', 'LiteracyRate', 'Total Schools', 'Teachers', 'Venue Counts']]
df
```

```python
df.sort_values('Population', ascending=False,inplace=True)
df
```

```python
df.reset_index(inplace=True)
df
```

```python
df.drop('index',axis=1,inplace=True)
df
```

```python
df = df[0:20]
df
```

```python
df.to_csv("ReallyFinalDF.csv")
```

```python
df.loc[1,'Venue Counts'] = 100
df
```

```python
df.to_csv("ReallyFinalDF.csv")
```

```python
X = df.values[:,3:].astype(float)
X_scaled = MinMaxScaler().fit_transform(X)
X_scaled
```

```python
num_clusters = 4

k_means = KMeans(init="k-means++", n_clusters=num_clusters, n_init=20)
k_means.fit(X_scaled)
labels = k_means.labels_

print(labels)
```

```python
df['ClusterLabels'] = labels
df
```

```python
df.to_csv("ReallyFinalDFwithClustering.csv")
```
