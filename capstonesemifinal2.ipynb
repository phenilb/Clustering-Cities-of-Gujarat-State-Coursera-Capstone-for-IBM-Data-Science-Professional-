```python
import pandas as pd
df = pd.read_pickle("citiesofgujarat.pickle")
df2 = pd.read_pickle("gujcitylatlng.pickle")
print(df.head())
print(df2.head())
```

```python
df.rename(columns={'Population 2011 Census':'Population',\
                   'Density 2011':'Population_Density'},inplace=True)
df
```

```python
df_withloc = pd.merge(df,df2, on='City', how='inner')
df_withloc
```

```python
df_withloc['Population'] = df_withloc['Population'].\
    replace({',': ''}, regex=True)
df_withloc
```

```python
df_withloc= df_withloc[df_withloc.Longitude != True]
df_withloc.reset_index(drop=True, inplace=True)
df_withloc
```
